```python
import torch
from untitled2 import Perceptron

def main():
  pytorch_model = Perceptron()
  pytorch_model.load_state_dict(torch.load('model.pt'))
  pytorch_model.eval()
  dummy_input = torch.zeros(1,2)
  torch.onnx.export(pytorch_model, dummy_input, 'model.onnx', verbose=True)

if __name__ == '__main__':
  main()
```
